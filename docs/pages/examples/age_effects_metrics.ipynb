```python
#import
import os
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt

from network_control.metrics import node_strength, ave_control, modal_control
from network_control.utils import rank_int, matrix_normalization
from network_control.plotting import set_plotting_params, reg_plot
set_plotting_params()
```

```python
# load data
workdir = '/Users/lindenmp/Google-Drive-Penn/work/research_projects/control_package/'
df = pd.read_csv(os.path.join(workdir, 'schaefer_200_streamlineCount_df.csv'))
# df = df.iloc[:100, :]

A = np.load(os.path.join(workdir, 'schaefer_200_streamlineCount_A.npy'))
# A = A[:, :, :100]
```

```python
print(A.shape)
```

```python
print(df['ageAtScan1'].head())
```

```python
n_nodes = A.shape[0]
n_subs = A.shape[2]
```

```python
# compute average and modal controllability
ac = np.zeros((n_subs, n_nodes))
mc = np.zeros((n_subs, n_nodes))

for i in np.arange(n_subs):
    # normalize subject's A matrix for a discrete-time system
    a_norm = matrix_normalization(A[:, :, i], version='discrete')
    ac[i, :] = ave_control(a_norm)
    mc[i, :] = modal_control(a_norm)
```

```python
# mean over nodes
ac_node_mean = np.mean(ac, axis=1)
mc_node_mean = np.mean(mc, axis=1)
```

## Plot

```python
f, ax = plt.subplots(1, 2, figsize=(5, 2.5))
reg_plot(x=df['ageAtScan1']/12, y=ac_node_mean, xlabel='Age (years)', ylabel='Average ctrb. (mean)', ax=ax[0])
reg_plot(x=df['ageAtScan1']/12, y=mc_node_mean, xlabel='Age (years)', ylabel='Modal ctrb. (mean)', ax=ax[1])
plt.show()
f.savefig(os.path.join(workdir, 'docs', 'pages', 'examples', 'age_effects_metrics'), dpi=300, bbox_inches='tight', pad_inches=0.1)
```
