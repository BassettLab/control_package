```python
#import

import numpy as np
import matplotlib.pyplot as plt

from network_control.metrics import node_strength, ave_control, modal_control
from network_control.utils import rank_int, matrix_normalization
from network_control.plotting import set_plotting_params, reg_plot
set_plotting_params()
```

```python
# load data
A = np.load('/Users/lindenmp/Google-Drive-Penn/work/research_projects/control_package/A.npy')
A = A[:, :, 0]
A_norm = matrix_normalization(A, version='discrete')
```

```python
# compute average and modal controllability
s = node_strength(A)
ac = ave_control(A_norm)
mc = modal_control(A_norm)
```

## Plot normalized values

```python
f, ax = plt.subplots(1, 3, figsize=(7.5, 2.5))
reg_plot(x=s, y=rank_int(ac), xlabel='strength', ylabel='average ctrb.', ax=ax[0])
reg_plot(x=s, y=rank_int(mc), xlabel='strength', ylabel='modal ctrb.', ax=ax[1])
reg_plot(x=rank_int(ac), y=rank_int(mc), xlabel='average ctrb.', ylabel='modal ctrb.', ax=ax[2])
plt.show()
f.savefig('corr_metrics', dpi=300, bbox_inches='tight', pad_inches=0.1)
```
