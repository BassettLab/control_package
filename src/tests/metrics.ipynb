```python
#import

import numpy as np
import matplotlib.pyplot as plt

from network_control.metrics import ave_control, modal_control
from network_control.utils import rank_int, matrix_normalization
from network_control.plotting import set_plotting_params, reg_plot
set_plotting_params()
```

```python
# load data
A = np.load('/Users/lindenmp/Google-Drive-Penn/work/research_projects/control_package/A.npy')
A = matrix_normalization(A[:, :, 0])
```

```python
# compute average and modal controllability
ac = ave_control(A)
mc = modal_control(A)
```

## Plot normalized values

```python
f, ax = plt.subplots(1, 1, figsize=(2.5, 2.5))
reg_plot(x=rank_int(ac), y=rank_int(mc), xlabel='average ctrb.', ylabel='modal ctrb.', ax=ax)
plt.show()
f.savefig('corr(ac,mc)', dpi=300, bbox_inches='tight', pad_inches=0.1)
```
